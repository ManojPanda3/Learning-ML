# Exponential Family
## PDF
$$ P(y;η) = b(y) \exp(\eta^TT(y)- a(\eta)) $$
-  y is Data
- η is natural parameter
- T(y) is sufficient statistics (here its y  the data it self was summary)
- b(y) is base measure
- a(η) is log partition


## Bernolli Distribution(Binary Data)

- $ \phi $ = probability of event happening or not
$$ P(y;ϕ) =  \phi^y (1-\phi)^{1-y} $$
$$ exp(log(\frac{\phi}{1-\phi})y + log(1-\phi))$$

here
- $ b(y) = 1 $
- $ η = log(\frac{\phi}{1-\phi}) ⇒ ϕ = \frac{1}{1+e^{-n}} $
- $ T(y) = y $
- $ a(η) = -log(1-ϕ) ⇒ -log(1-\frac{1}{1+e^{-n}}) = log(1 + e^n) $

# Gaussian (with fixed variance)

Assume $\sigma^2 = 1$
$$P(y;\mu) = \frac{1}{\sqrt{2\pi}} exp(-\frac{(y - \mu)^2}{2}) $$
$$P(y;\mu) = \frac{1}{\sqrt{2\pi}}e^{-y^2/2} exp(\mu y - 1/2 \mu^2) $$

- $b(y) = \frac{1}{\sqrt{2\pi}}e^{-y^2/2}$
- $\eta = \mu$
- $T(y) = y$
- $a(\eta)=1/2\mu^2$
