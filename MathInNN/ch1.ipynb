# MachineLearning

A machine is said to be learning if it does task T and gain experiance E with a performance P and the P increaces over time as more and more E was gathered by the computer.

Theres mainly 2 type of Learning Algorithms/Techniques
## 1. Supervised Learning
A Technique where data are labeled and machine is trained to match the output of already labeled data.

Theres 2 type of Supervised Learning: 
1. Regression
2. Classification

## 2. Unsupervised Learning
A technique where we ask the ai to do task T and according to the perfomance P we give it feedback F it can be good feedback or bad feedback according to the P. And the Ai model try to get the optimal answer by chessing the Good feedback F.
