$$J(W) = \frac{1}{2}\sum_{i=1}^{m}(h_W(x^{(i)}) - y^{(i)})^2$$

## Gradient Descent

start with some value of W like W=$\vec{0}$
We will then keep changing the W to minize the $J(W)$
$$W_j := W_j - \alpha \frac{\partial J(W)}{\partial W_j}$$

1. $\alpha$ is learning rate.

### Batch gradient decent

$$\frac{\partial J(W)}{\partial W_j} = \sum_{i=0}^{m}{(h_W(x^i) - y^i) x^i_j}$$
Now $W_j := W_j - \alpha  \sum_{i=1}^{m}{(h_W(x^i) - y^i) x^i_j}$

This is called batch gradient descent because it proccesses batches of data.

### Stochastic gradient decent

Here we go through all the dataset and adjust each weight for that dataset one by one

```python
def batch_gradient():
    W = []; # parameters
    alpha = 0.01; # learning rate
    x = [[]]; # input/features
    y = [[]]; # target

    # looping through the weights
    for j in range(len(W)):        
        W[j] = W[j] - alpha * sum((h(x[i])-y[i]) * x[i] for i in range(m)); # optimizing current weight for all the dataset
```

```python
def stochastic_gradient():
    """
    here for each weight in the Neural layer we are going through all the rows of data set and adjusting that weight so that it can suit to all the datas
    """
    W = []; # parameters
    alpha = 0.01; # learning rate
    x = [[]]; # input/features
    y = [[]]; # target

    # looping through the weights
    for j in range(len(W)):
        # looping through all the datas
        for i in range(1,m):
            W[j] = W[j] - alpha * (h(x[i])-y[i]) * x[i]; # optimizing current weight for the current dataset
```
